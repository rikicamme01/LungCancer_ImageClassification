```python
!pip install pynrrd

import os
import random
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras import models, layers
import nrrd  # Make sure this is installed in your environment
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.metrics import confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns

################################################################################
# Set a random seed for reproducibility:
seed = 42
random.seed(seed)
np.random.seed(seed)
tf.random.set_seed(seed)
################################################################################

excel_path = "/kaggle/input/my-lung-dataset/my_lung_dataset/dataset_lung.xlsx"
data_dir = "/kaggle/input/my-lung-dataset/my_lung_dataset/Train"

df = pd.read_excel(excel_path)

nodule_paths = []
labels_binary = []

for idx, row in df.iterrows():
    nodule_filename = row["Nodule"]       # e.g., "pat1_nodule.nrrd"
    tumor_class = row["TumorClass"]       # e.g. 1..5
    full_nodule_path = os.path.join(data_dir, nodule_filename)
    
    # Convert classes 1,2,3 => 0 (benign), 4,5 => 1 (malignant)
    label_bin = 0 if tumor_class in [1,2,3] else 1
    
    nodule_paths.append(full_nodule_path)
    labels_binary.append(label_bin)

data_nodules = pd.DataFrame({
    "nodule_path": nodule_paths,
    "label": labels_binary
})

print("Total nodule images:", len(data_nodules))
display(data_nodules.head())

################################################################################
# Split into train+val (80%) and final test (20%)
################################################################################
trainval_df, test_df = train_test_split(
    data_nodules, 
    test_size=0.20,
    stratify=data_nodules["label"], 
    random_state=seed  # keep same seed for reproducibility
)

print("Train+Val samples:", len(trainval_df))
print("Final Test samples:", len(test_df))

################################################################################
# Image loading and tf.data Dataset creation
################################################################################
IMG_SIZE = (128, 128)

def load_nrrd_image(nrrd_path, target_size=IMG_SIZE):
    volume_data, header = nrrd.read(nrrd_path)

    # If 3D, reduce to 2D if needed
    if volume_data.ndim == 3:
        if volume_data.shape[-1] == 1:
            volume_data = volume_data[..., 0]
        elif volume_data.shape[0] == 1:
            volume_data = volume_data[0, ...]

    volume_data = volume_data.astype(np.float32)
    volume_data = tf.image.resize(volume_data[..., tf.newaxis], target_size)
    volume_data = tf.image.grayscale_to_rgb(volume_data)
    return volume_data.numpy()

def create_dataset(df, batch_size=8, shuffle=False):
    paths = df["nodule_path"].tolist()
    labels = df["label"].tolist()
    
    def gen():
        for p, l in zip(paths, labels):
            img = load_nrrd_image(p)
            yield img, l
    
    dataset = tf.data.Dataset.from_generator(
        gen,
        output_types=(tf.float32, tf.int32),
        output_shapes=((IMG_SIZE[0], IMG_SIZE[1], 3), ())
    )
    if shuffle:
        dataset = dataset.shuffle(buffer_size=len(df), seed=seed)  # seed for dataset shuffle
    dataset = dataset.batch(batch_size).prefetch(tf.data.AUTOTUNE)
    return dataset

################################################################################
# Simple CNN model definition
################################################################################
def create_cnn_model(input_shape=(128,128,3)):
    model = models.Sequential()
    model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=input_shape))
    model.add(layers.MaxPooling2D((2, 2)))
    model.add(layers.Conv2D(64, (3, 3), activation='relu'))
    model.add(layers.MaxPooling2D((2, 2)))
    model.add(layers.Conv2D(128, (3, 3), activation='relu'))
    model.add(layers.MaxPooling2D((2, 2)))

    model.add(layers.Flatten())
    model.add(layers.Dense(128, activation='relu'))
    model.add(layers.Dropout(0.3))
    model.add(layers.Dense(1, activation='sigmoid'))

    model.compile(optimizer='adam', 
                  loss='binary_crossentropy', 
                  metrics=['accuracy'])
    return model

################################################################################
# Cross validation on trainval_df
################################################################################
N_FOLDS = 5
BATCH_SIZE = 8
EPOCHS = 30

skf = StratifiedKFold(n_splits=N_FOLDS, shuffle=True, random_state=seed)

X = trainval_df["nodule_path"].values
y = trainval_df["label"].values

fold_accuracies = []
fold_conf_matrices = []

# Callbacks
patience_es = 5
patience_lr = 3
callbacks = [
    tf.keras.callbacks.EarlyStopping(
        patience=patience_es, restore_best_weights=True, 
        monitor='val_loss', mode='min'
    ),
    tf.keras.callbacks.ReduceLROnPlateau(
        factor=0.1, patience=patience_lr, min_lr=1e-5,
        monitor='val_loss', mode='min'
    )
]

fold_num = 1
for train_index, val_index in skf.split(X, y):
    print(f"\n=== Fold {fold_num}/{N_FOLDS} ===")
    fold_num += 1
    
    # Build dataframes for this fold
    df_train_fold = trainval_df.iloc[train_index].reset_index(drop=True)
    df_val_fold   = trainval_df.iloc[val_index].reset_index(drop=True)
    
    # Optional: compute class weights if imbalanced
    train_labels_value_counts = df_train_fold["label"].value_counts()
    benign_count = train_labels_value_counts.get(0, 0)
    malignant_count = train_labels_value_counts.get(1, 0)
    total = benign_count + malignant_count
    weight_for_0 = (1 / benign_count) * (total / 2.0)
    weight_for_1 = (1 / malignant_count) * (total / 2.0)
    class_weights = {0: weight_for_0, 1: weight_for_1}
    
    # Create datasets
    ds_train_fold = create_dataset(df_train_fold, batch_size=BATCH_SIZE, shuffle=True)
    ds_val_fold   = create_dataset(df_val_fold,   batch_size=BATCH_SIZE, shuffle=False)

    # Create a fresh model
    model = create_cnn_model(input_shape=(IMG_SIZE[0], IMG_SIZE[1], 3))

    # Train
    history = model.fit(
        ds_train_fold,
        epochs=EPOCHS,
        validation_data=ds_val_fold,
        callbacks=callbacks,
        class_weight=class_weights,
        verbose=1
    )

    # Evaluate
    val_loss, val_acc = model.evaluate(ds_val_fold, verbose=0)
    print(f"Validation Loss: {val_loss:.4f}, Validation Accuracy: {val_acc:.4f}")
    
    # Collect predictions for confusion matrix
    val_preds = model.predict(ds_val_fold).ravel()
    val_preds = (val_preds > 0.5).astype(int)

    # True labels
    val_true = []
    for _, lbl_batch in ds_val_fold:
        val_true.extend(lbl_batch.numpy())

    cm = confusion_matrix(val_true, val_preds)
    print("Confusion Matrix (Val):\n", cm)
    print(classification_report(val_true, val_preds, 
                                target_names=["Benign (0)", "Malignant (1)"]))
    
    fold_accuracies.append(val_acc)
    fold_conf_matrices.append(cm)

# After cross validation
mean_acc = np.mean(fold_accuracies)
std_acc  = np.std(fold_accuracies)
print("\n=== Cross Validation Results ===")
print(f"Mean Val Accuracy: {mean_acc:.4f} ± {std_acc:.4f}")

################################################################################
# 7) Retrain on All Train+Val data with final hyperparams, then Test on hold-out
################################################################################
FINAL_EPOCHS = 20
FINAL_BATCH_SIZE = 8

final_callbacks = [
    tf.keras.callbacks.EarlyStopping(
        patience=patience_es, 
        restore_best_weights=True,
        monitor='loss', 
        mode='min'
    ),
    tf.keras.callbacks.ReduceLROnPlateau(
        factor=0.1, patience=patience_lr, min_lr=1e-5,
        monitor='loss', 
        mode='min'
    )
]

# Create dataset from the entire trainval_df
ds_trainval = create_dataset(trainval_df, batch_size=FINAL_BATCH_SIZE, shuffle=True)

# Class weights for the entire trainval set
trainval_labels = trainval_df["label"].value_counts()  # 0 or 1
benign_count = trainval_labels.get(0, 0)
malignant_count = trainval_labels.get(1, 0)
total = benign_count + malignant_count
weight_for_0 = (1 / benign_count) * (total / 2.0)
weight_for_1 = (1 / malignant_count) * (total / 2.0)
class_weights_final = {0: weight_for_0, 1: weight_for_1}
print("\nFinal class weights:", class_weights_final)

# Build fresh model
final_model = create_cnn_model(input_shape=(IMG_SIZE[0], IMG_SIZE
                                            [1], 3))

# Train on entire trainval_df
final_history = final_model.fit(
    ds_trainval,
    epochs=FINAL_EPOCHS,
    callbacks=final_callbacks,
    class_weight=class_weights_final,
    verbose=1
)

# Evaluate on hold-out test
ds_test = create_dataset(test_df, batch_size=FINAL_BATCH_SIZE, shuffle=False)
test_loss, test_acc = final_model.evaluate(ds_test, verbose=0)
print(f"\n** Final Test Loss: {test_loss:.4f}, Final Test Accuracy: {test_acc:.4f}")

test_preds = final_model.predict(ds_test).ravel()
test_preds = (test_preds > 0.5).astype(int)

test_labels_list = []
for _, lbl_batch in ds_test:
    test_labels_list.extend(lbl_batch.numpy())

cm_test = confusion_matrix(test_labels_list, test_preds)
print("\nHold-Out Test Confusion Matrix:\n", cm_test)
print("Hold-Out Classification Report:\n",
      classification_report(test_labels_list, test_preds,
                            target_names=["Benign (0)", "Malignant (1)"]))

# OPTIONAL: PLOT TEST CONFUSION MATRIX
plt.figure(figsize=(5,4))
sns.heatmap(cm_test,
            annot=True, cmap='Blues', fmt='d',
            xticklabels=["Benign (0)", "Malignant (1)"],
            yticklabels=["Benign (0)", "Malignant (1)"])
plt.title("Confusion Matrix - Test Set")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()
```

```python
import cv2
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

def compute_gradcam(model, img_array, layer_name):
    """
    Compute Grad-CAM heatmap for a given image.
    
    Parameters:
    - model: Trained CNN model
    - img_array: Preprocessed image array (shape: (128, 128, 3))
    - layer_name: Name of the last convolutional layer to extract feature maps
    
    Returns:
    - Heatmap (128x128) of Grad-CAM
    """
    grad_model = tf.keras.models.Model(
        inputs=model.input,
        outputs=[model.get_layer(layer_name).output, model.output]
    )
    
    with tf.GradientTape() as tape:
        conv_outputs, predictions = grad_model(np.expand_dims(img_array, axis=0))
        class_idx = int(predictions > 0.5)  # 0 = benign, 1 = malignant
        loss = predictions[:, class_idx]
    
    grads = tape.gradient(loss, conv_outputs)
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))
    
    conv_outputs = conv_outputs[0]
    heatmap = tf.reduce_sum(pooled_grads * conv_outputs, axis=-1)
    heatmap = np.maximum(heatmap, 0)  # ReLU to remove negative values
    heatmap /= np.max(heatmap)  # Normalize between 0 and 1
    return heatmap.numpy()

def superimpose_gradcam(img, heatmap, alpha=0.4):
    """Overlay heatmap onto original image."""
    heatmap = cv2.resize(heatmap, (img.shape[1], img.shape[0]))
    heatmap = np.uint8(255 * heatmap)
    heatmap = cv2.applyColorMap(heatmap, cv2.COLORMAP_JET)
    
    superimposed_img = cv2.addWeighted(img, 1 - alpha, heatmap, alpha, 0)
    return superimposed_img

def plot_gradcam(model, img_path, layer_name="conv2d_2"):
    """
    Load an image, compute Grad-CAM, and display the results.
    
    Parameters:
    - model: Trained CNN model
    - img_path: Path to the NRRD image
    - layer_name: Layer to extract Grad-CAM
    """
    img_array = load_nrrd_image(img_path)
    heatmap = compute_gradcam(model, img_array, layer_name)
    overlay_img = superimpose_gradcam(np.uint8(img_array * 255), heatmap)
    
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 3, 1)
    plt.imshow(img_array)
    plt.title("Original Image")
    
    plt.subplot(1, 3, 2)
    plt.imshow(heatmap, cmap='jet')
    plt.title("Grad-CAM Heatmap")
    
    plt.subplot(1, 3, 3)
    plt.imshow(overlay_img)
    plt.title("Superimposed Grad-CAM")
    plt.show()

# Esempio d'uso su un'immagine casuale del test set
sample_img_path = test_df.iloc[0]['nodule_path']
plot_gradcam(final_model, sample_img_path, layer_name="conv2d_2")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

def get_gradcam_heatmap(img_array, model, last_conv_layer_name):
    """
    Data un'immagine (con batch size = 1), restituisce la heatmap Grad-CAM.
    """
    # Forza la chiamata del modello con un input dummy per assicurarsi che il grafo sia costruito
    dummy_input = tf.zeros((1, IMG_SIZE[0], IMG_SIZE[1], 3))
    _ = model(dummy_input)
    
    # Costruisci un modello intermedio che restituisce l'output dell'ultimo layer convoluzionale e quello finale
    grad_model = tf.keras.models.Model(
        inputs=model.inputs,
        outputs=[model.get_layer(last_conv_layer_name).output, model.output]
    )
    
    with tf.GradientTape() as tape:
        conv_outputs, predictions = grad_model(img_array)
        # Per classificazione binaria con attivazione sigmoid la prediction è uno scalare
        loss = predictions[0]
    
    # Calcola il gradiente della loss rispetto all'output del layer target
    grads = tape.gradient(loss, conv_outputs)
    
    # Esegui il global average pooling dei gradienti per ottenere il peso di ciascun filtro
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))
    
    # Rimuovi la dimensione batch dalle attivazioni
    conv_outputs = conv_outputs[0]
    
    # Moltiplica ogni feature map per il peso corrispondente
    heatmap = conv_outputs * pooled_grads
    # Somma lungo l'asse dei canali per ottenere una mappa 2D
    heatmap = tf.reduce_sum(heatmap, axis=-1)
    
    # Applica ReLU e normalizza la heatmap in [0, 1]
    heatmap = tf.maximum(heatmap, 0)
    max_val = tf.reduce_max(heatmap)
    if max_val == 0:
        max_val = tf.constant(1e-10)
    heatmap = heatmap / max_val
    return heatmap.numpy()

def superimpose_heatmap(img, heatmap, alpha=0.4, colormap=cv2.COLORMAP_JET):
    """
    Sovrappone la heatmap all'immagine originale.
    """
    # Ridimensiona la heatmap alle dimensioni dell'immagine
    heatmap = cv2.resize(heatmap, (img.shape[1], img.shape[0]))
    # Porta la heatmap in scala 0-255 e applica il colormap JET
    heatmap_uint8 = np.uint8(255 * heatmap)
    heatmap_color = cv2.applyColorMap(heatmap_uint8, colormap)
    
    # Se l'immagine non è in uint8 (supponendo che i pixel siano in [0,1]), converti
    if img.dtype != np.uint8:
        img_uint8 = np.uint8(255 * img)
    else:
        img_uint8 = img.copy()
    
    # Combina immagine e heatmap con trasparenza alpha
    overlay = cv2.addWeighted(img_uint8, 1 - alpha, heatmap_color, alpha, 0)
    return overlay

# ------------------ ESECUZIONE: CALCOLO DELLA GRAD-CAM ------------------

# Seleziona una immagine dal test set (prendi il primo batch e il primo campione)
for images, labels in ds_test.take(1):
    test_img = images[0].numpy()  # Dimensione: (128, 128, 3)
    break

# Individua il nome dell'ultimo layer convoluzionale. 
# Nel nostro modello Sequential, il layer a indice 4 (Conv2D con 128 filtri) è il target.
last_conv_layer_name = final_model.layers[4].name
print("Utilizzo il layer:", last_conv_layer_name)

# Prepara l'immagine per il modello (aggiungi dimensione batch)
img_array = np.expand_dims(test_img, axis=0)

# Calcola la Grad-CAM heatmap
heatmap = get_gradcam_heatmap(img_array, final_model, last_conv_layer_name)

# Crea l'immagine con la heatmap sovrapposta
overlay_img = superimpose_heatmap(test_img, heatmap, alpha=0.4)

# Visualizza l'immagine originale e quella con Grad-CAM affiancate
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

axs[0].imshow(test_img)
axs[0].set_title("Immagine di Test Originale")
axs[0].axis("off")

axs[1].imshow(overlay_img)
axs[1].set_title("Overlay Grad-CAM")
axs[1].axis("off")

plt.show()
```

```python
!pip install pynrrd

import os
import random
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras import models, layers
import nrrd  # Make sure this is installed
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.metrics import confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns

################################################################################
# Set a random seed for reproducibility:
seed = 42
random.seed(seed)
np.random.seed(seed)
tf.random.set_seed(seed)
################################################################################

excel_path = "/kaggle/input/my-lung-dataset/my_lung_dataset/dataset_lung.xlsx"
data_dir = "/kaggle/input/my-lung-dataset/my_lung_dataset/Train"

df = pd.read_excel(excel_path)

nodule_paths = []
labels_binary = []

for idx, row in df.iterrows():
    nodule_filename = row["Nodule"]       # e.g., "pat1_nodule.nrrd"
    tumor_class = row["TumorClass"]       # e.g. 1..5
    full_nodule_path = os.path.join(data_dir, nodule_filename)
    
    # Convert classes 1,2,3 => 0 (benign), 4,5 => 1 (malignant)
    label_bin = 0 if tumor_class in [1,2,3] else 1
    
    nodule_paths.append(full_nodule_path)
    labels_binary.append(label_bin)

data_nodules = pd.DataFrame({
    "nodule_path": nodule_paths,
    "label": labels_binary
})

print("Total nodule images:", len(data_nodules))
display(data_nodules.head())

################################################################################
# Split into train+val (80%) and final test (20%)
################################################################################
trainval_df, test_df = train_test_split(
    data_nodules, 
    test_size=0.20,
    stratify=data_nodules["label"], 
    random_state=seed
)

print("Train+Val samples:", len(trainval_df))
print("Final Test samples:", len(test_df))

################################################################################
# 1) Data Augmentation Function (no TFA)
################################################################################

def augment_image(image):
    """
    Apply random flips, random 90-degree rotation, random brightness, etc.
    Only used for malignant samples (oversampling the minority class).
    """
    # Convert to Tensor
    x = tf.convert_to_tensor(image, dtype=tf.float32)
    
    # Random flips
    x = tf.image.random_flip_left_right(x)
    x = tf.image.random_flip_up_down(x)
    
    # Random 90-degree rotation (k in {0,1,2,3})
    k = tf.random.uniform([], minval=0, maxval=4, dtype=tf.int32)
    x = tf.image.rot90(x, k)

    # Random brightness
    x = tf.image.random_brightness(x, max_delta=0.1)

    return x.numpy()  # return as NumPy array

################################################################################
# 2) Define image loader & dataset creation with optional minority augmentation
################################################################################

IMG_SIZE = (128, 128)

def load_nrrd_image(nrrd_path, target_size=IMG_SIZE):
    volume_data, header = nrrd.read(nrrd_path)
    if volume_data.ndim == 3:
        if volume_data.shape[-1] == 1:
            volume_data = volume_data[..., 0]
        elif volume_data.shape[0] == 1:
            volume_data = volume_data[0, ...]

    volume_data = volume_data.astype(np.float32)
    volume_data = tf.image.resize(volume_data[..., tf.newaxis], target_size)
    volume_data = tf.image.grayscale_to_rgb(volume_data)
    return volume_data.numpy()

def create_dataset(df, batch_size=8, shuffle=False, augment_minority=False, augment_factor=1):
    """
    If augment_minority=True, each malignant (label=1) sample is duplicated 
    'augment_factor' times with random augmentations.
    """
    paths = df["nodule_path"].tolist()
    labels = df["label"].tolist()
    
    def gen():
        for p, l in zip(paths, labels):
            # Always yield the original
            img = load_nrrd_image(p, target_size=IMG_SIZE)
            yield img, l

            # If malignant & augmentation is ON, yield extra augmented copies
            if augment_minority and l == 1:
                for _ in range(augment_factor):
                    aug_img = augment_image(img)
                    yield aug_img, l

    dataset = tf.data.Dataset.from_generator(
        gen,
        output_types=(tf.float32, tf.int32),
        output_shapes=((IMG_SIZE[0], IMG_SIZE[1], 3), ())
    )
    if shuffle:
        dataset = dataset.shuffle(buffer_size=len(df), seed=seed)
    dataset = dataset.batch(batch_size).prefetch(tf.data.AUTOTUNE)
    return dataset

################################################################################
# 3) Simple CNN model definition (same as your original)
################################################################################

def create_cnn_model(input_shape=(128,128,3)):
    model = models.Sequential()
    model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=input_shape))
    model.add(layers.MaxPooling2D((2, 2)))
    model.add(layers.Conv2D(64, (3, 3), activation='relu'))
    model.add(layers.MaxPooling2D((2, 2)))
    model.add(layers.Conv2D(128, (3, 3), activation='relu'))
    model.add(layers.MaxPooling2D((2, 2)))

    model.add(layers.Flatten())
    model.add(layers.Dense(128, activation='relu'))
    model.add(layers.Dropout(0.3))
    model.add(layers.Dense(1, activation='sigmoid'))

    model.compile(optimizer='adam', 
                  loss='binary_crossentropy', 
                  metrics=['accuracy'])
    return model

################################################################################
# 4) Cross validation on trainval_df (with augmentation on the minority class)
################################################################################

N_FOLDS = 5
BATCH_SIZE = 8
EPOCHS = 30
augment_factor = 1  # how many extra augmented copies for each malignant sample

skf = StratifiedKFold(n_splits=N_FOLDS, shuffle=True, random_state=seed)

X = trainval_df["nodule_path"].values
y = trainval_df["label"].values

fold_accuracies = []
fold_conf_matrices = []

# Callbacks
patience_es = 5
patience_lr = 3
callbacks = [
    tf.keras.callbacks.EarlyStopping(
        patience=patience_es, restore_best_weights=True, 
        monitor='val_loss', mode='min'
    ),
    tf.keras.callbacks.ReduceLROnPlateau(
        factor=0.1, patience=patience_lr, min_lr=1e-5,
        monitor='val_loss', mode='min'
    )
]

fold_num = 1
for train_index, val_index in skf.split(X, y):
    print(f"\n=== Fold {fold_num}/{N_FOLDS} ===")
    
    df_train_fold = trainval_df.iloc[train_index].reset_index(drop=True)
    df_val_fold   = trainval_df.iloc[val_index].reset_index(drop=True)
    
    # Compute class weights if needed
    train_counts = df_train_fold["label"].value_counts()
    benign_count = train_counts.get(0, 0)
    malignant_count = train_counts.get(1, 0)
    total = benign_count + malignant_count
    weight_for_0 = (1 / benign_count) * (total / 2.0)
    weight_for_1 = (1 / malignant_count) * (total / 2.0)
    class_weights = {0: weight_for_0, 1: weight_for_1}
    
    # Create datasets
    # Notice: augment_minority=True only for training fold
    ds_train_fold = create_dataset(df_train_fold, 
                                   batch_size=BATCH_SIZE, 
                                   shuffle=True, 
                                   augment_minority=True, 
                                   augment_factor=augment_factor)
    ds_val_fold   = create_dataset(df_val_fold, 
                                   batch_size=BATCH_SIZE, 
                                   shuffle=False, 
                                   augment_minority=False)

    model = create_cnn_model(input_shape=(IMG_SIZE[0], IMG_SIZE[1], 3))

    # Train
    history = model.fit(
        ds_train_fold,
        epochs=EPOCHS,
        validation_data=ds_val_fold,
        callbacks=callbacks,
        class_weight=class_weights,
        verbose=1
    )

    # Evaluate on val fold
    val_loss, val_acc = model.evaluate(ds_val_fold, verbose=0)
    print(f"Fold {fold_num} - Val Loss: {val_loss:.4f}, Val Accuracy: {val_acc:.4f}")
    
    # Confusion matrix
    val_preds = model.predict(ds_val_fold).ravel()
    val_preds = (val_preds > 0.5).astype(int)

    val_true = []
    for _, lbl_batch in ds_val_fold:
        val_true.extend(lbl_batch.numpy())

    cm = confusion_matrix(val_true, val_preds)
    print("Confusion Matrix (Val):\n", cm)
    print(classification_report(val_true, val_preds,
          target_names=["Benign (0)", "Malignant (1)"]))
    
    fold_accuracies.append(val_acc)
    fold_conf_matrices.append(cm)
    fold_num += 1

mean_acc = np.mean(fold_accuracies)
std_acc  = np.std(fold_accuracies)
print("\n=== Cross Validation Results ===")
print(f"Mean Val Accuracy: {mean_acc:.4f} ± {std_acc:.4f}")

################################################################################
# 5) Retrain on entire trainval_df (with augmentation) & Evaluate on test set
################################################################################

FINAL_EPOCHS = 20
FINAL_BATCH_SIZE = 8

final_callbacks = [
    tf.keras.callbacks.EarlyStopping(
        patience=patience_es, 
        restore_best_weights=True,
        monitor='loss', 
        mode='min'
    ),
    tf.keras.callbacks.ReduceLROnPlateau(
        factor=0.1, patience=patience_lr, min_lr=1e-5,
        monitor='loss', 
        mode='min'
    )
]

ds_trainval = create_dataset(trainval_df, 
                             batch_size=FINAL_BATCH_SIZE, 
                             shuffle=True, 
                             augment_minority=True, 
                             augment_factor=augment_factor)

trainval_counts = trainval_df["label"].value_counts()
benign_count = trainval_counts.get(0, 0)
malignant_count = trainval_counts.get(1, 0)
total = benign_count + malignant_count
weight_for_0 = (1 / benign_count) * (total / 2.0)
weight_for_1 = (1 / malignant_count) * (total / 2.0)
class_weights_final = {0: weight_for_0, 1: weight_for_1}
print("\nFinal class weights:", class_weights_final)

final_model = create_cnn_model(input_shape=(IMG_SIZE[0], IMG_SIZE[1], 3))

final_history = final_model.fit(
    ds_trainval,
    epochs=FINAL_EPOCHS,
    callbacks=final_callbacks,
    class_weight=class_weights_final,
    verbose=1
)

ds_test = create_dataset(test_df, batch_size=FINAL_BATCH_SIZE, shuffle=False, augment_minority=False)
test_loss, test_acc = final_model.evaluate(ds_test, verbose=0)
print(f"\n** Final Test Loss: {test_loss:.4f}, Final Test Accuracy: {test_acc:.4f}")

test_preds = final_model.predict(ds_test).ravel()
test_preds = (test_preds > 0.5).astype(int)

test_labels_list = []
for _, lbl_batch in ds_test:
    test_labels_list.extend(lbl_batch.numpy())

cm_test = confusion_matrix(test_labels_list, test_preds)
print("\nHold-Out Test Confusion Matrix:\n", cm_test)
print("Hold-Out Classification Report:\n",
      classification_report(test_labels_list, test_preds,
                            target_names=["Benign (0)", "Malignant (1)"]))

plt.figure(figsize=(5,4))
sns.heatmap(cm_test,
            annot=True, cmap='Blues', fmt='d',
            xticklabels=["Benign (0)", "Malignant (1)"],
            yticklabels=["Benign (0)", "Malignant (1)"])
plt.title("Confusion Matrix - Test Set")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()
```
