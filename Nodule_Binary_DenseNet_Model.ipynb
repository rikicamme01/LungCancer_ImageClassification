```python
!pip install pynrrd

import os
import random
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras import models, layers
import nrrd
from sklearn.model_selection import StratifiedKFold, train_test_split
from sklearn.metrics import confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns

################################################################################
# 0) Seed initialization for reproducibility
################################################################################
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)

################################################################################
# 1) Load the Excel file and create the DataFrame
################################################################################

excel_path = "/kaggle/input/my-lung-dataset/my_lung_dataset/dataset_lung.xlsx"
data_dir = "/kaggle/input/my-lung-dataset/my_lung_dataset/Train"

df = pd.read_excel(excel_path)

nodule_paths = []
labels_binary = []

for idx, row in df.iterrows():
    nodule_filename = row["Nodule"]       # e.g. "pat1_nodule.nrrd"
    tumor_class = row["TumorClass"]       # e.g. 1..5
    full_nodule_path = os.path.join(data_dir, nodule_filename)
    
    # [1,2,3] => 0 (benign), [4,5] => 1 (malignant)
    if tumor_class in [1,2,3]:
        label_bin = 0
    else:
        label_bin = 1
    
    nodule_paths.append(full_nodule_path)
    labels_binary.append(label_bin)

data_nodules = pd.DataFrame({
    "nodule_path": nodule_paths,
    "label": labels_binary
})

print("Total nodule images:", len(data_nodules))
display(data_nodules.head())

################################################################################
# 1.5) Split off a final hold-out test set (e.g., 20%)
################################################################################

trainval_df, test_df = train_test_split(
    data_nodules, 
    test_size=0.20,
    stratify=data_nodules["label"],
    random_state=SEED
)

print("\nTrain+Val samples:", len(trainval_df))
print("Test samples:", len(test_df))

################################################################################
# 2) Define image loading & dataset creation
################################################################################

IMG_SIZE = (128, 128)

def load_nrrd_image(nrrd_path, target_size=IMG_SIZE):
    volume_data, header = nrrd.read(nrrd_path)
    
    if volume_data.ndim == 3:
        if volume_data.shape[-1] == 1:
            volume_data = volume_data[..., 0]
        elif volume_data.shape[0] == 1:
            volume_data = volume_data[0, ...]
    
    volume_data = volume_data.astype(np.float32)
    volume_data = tf.image.resize(volume_data[..., tf.newaxis], target_size)
    
    # Convert to 3 channels (for pretrained networks):
    volume_data = tf.image.grayscale_to_rgb(volume_data)
    return volume_data.numpy()

def create_dataset(df, batch_size=8, shuffle=False):
    paths = df["nodule_path"].tolist()
    labels = df["label"].tolist()
    
    def gen():
        for p, l in zip(paths, labels):
            img = load_nrrd_image(p, target_size=IMG_SIZE)
            yield img, l

    dataset = tf.data.Dataset.from_generator(
        gen,
        output_types=(tf.float32, tf.int32),
        output_shapes=((IMG_SIZE[0], IMG_SIZE[1], 3), ())
    )
    if shuffle:
        dataset = dataset.shuffle(buffer_size=len(df), seed=SEED)
    dataset = dataset.batch(batch_size).prefetch(tf.data.AUTOTUNE)
    return dataset

################################################################################
# 3) Create a Transfer Learning Model (DenseNet + Fine Tuning)
################################################################################

def create_transfer_model(input_shape=(128,128,3), fine_tune_at=None):
    """
    Builds a model with a DenseNet121 backbone (pretrained on ImageNet).
    If `fine_tune_at` is None, we freeze the entire base model (only train the new head).
    If `fine_tune_at` is an integer, we'll unfreeze layers after that index for fine-tuning.
    """
    
    base_model = tf.keras.applications.DenseNet121(
        input_shape=input_shape,
        include_top=False,
        weights='imagenet'
    )
    
    if fine_tune_at is None:
        base_model.trainable = False
    else:
        # Unfreeze from 'fine_tune_at' layer onward
        for layer in base_model.layers[:fine_tune_at]:
            layer.trainable = False
        for layer in base_model.layers[fine_tune_at:]:
            layer.trainable = True
    
    model = tf.keras.Sequential([
        base_model,
        layers.GlobalAveragePooling2D(),
        layers.Dense(128, activation='relu'),
        layers.Dropout(0.3),
        layers.Dense(1, activation='sigmoid')
    ])
    
    model.compile(optimizer='adam',
                  loss='binary_crossentropy',
                  metrics=['accuracy'])
    
    return model

################################################################################
# 4) Perform Cross Validation on trainval_df (StratifiedKFold)
################################################################################

N_FOLDS = 5
BATCH_SIZE = 8
EPOCHS = 10  # You can adjust

from sklearn.model_selection import StratifiedKFold

skf = StratifiedKFold(n_splits=N_FOLDS, shuffle=True, random_state=SEED)

X = trainval_df["nodule_path"].values
y = trainval_df["label"].values

fold_accuracies = []
fold_conf_matrices = []

# Callbacks
patience_es = 5
patience_lr = 3
callbacks = [
    tf.keras.callbacks.EarlyStopping(
        patience=patience_es, 
        restore_best_weights=True, 
        monitor='val_loss', 
        mode='min'
    ),
    tf.keras.callbacks.ReduceLROnPlateau(
        factor=0.1, 
        patience=patience_lr, 
        min_lr=1e-5,
        monitor='val_loss',
        mode='min'
    )
]

fold_num = 1
for train_index, val_index in skf.split(X, y):
    print(f"\n=== Fold {fold_num}/{N_FOLDS} ===")
    
    df_train_fold = trainval_df.iloc[train_index].reset_index(drop=True)
    df_val_fold   = trainval_df.iloc[val_index].reset_index(drop=True)
    
    # Optional: compute class weights for this fold
    train_counts = df_train_fold["label"].value_counts()
    benign_count = train_counts.get(0, 0)
    malignant_count = train_counts.get(1, 0)
    total = benign_count + malignant_count
    weight_for_0 = (1 / benign_count) * (total / 2.0)
    weight_for_1 = (1 / malignant_count) * (total / 2.0)
    fold_class_weights = {0: weight_for_0, 1: weight_for_1}
    
    ds_train_fold = create_dataset(df_train_fold, batch_size=BATCH_SIZE, shuffle=True)
    ds_val_fold   = create_dataset(df_val_fold,   batch_size=BATCH_SIZE, shuffle=False)
    
    # Create the transfer model
    transfer_model = create_transfer_model(input_shape=(128,128,3), fine_tune_at=100)
    
    # Train
    history = transfer_model.fit(
        ds_train_fold,
        epochs=EPOCHS,
        validation_data=ds_val_fold,
        callbacks=callbacks,
        class_weight=fold_class_weights,
        verbose=1
    )
    
    # Evaluate on this fold's validation set
    val_loss, val_acc = transfer_model.evaluate(ds_val_fold, verbose=0)
    print(f"Fold {fold_num} - Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.4f}")
    
    # Predictions & confusion matrix
    val_preds = transfer_model.predict(ds_val_fold).ravel()
    val_preds = (val_preds > 0.5).astype(int)
    
    val_true = []
    for _, lbl_batch in ds_val_fold:
        val_true.extend(lbl_batch.numpy())
    
    cm = confusion_matrix(val_true, val_preds)
    fold_conf_matrices.append(cm)
    
    print("Confusion Matrix:\n", cm)
    print(classification_report(val_true, val_preds,
          target_names=["Benign (0)", "Malignant (1)"]))
    
    fold_accuracies.append(val_acc)
    fold_num += 1

print("\n=== Cross Validation Results ===")
mean_acc = np.mean(fold_accuracies)
std_acc  = np.std(fold_accuracies)
print(f"Mean Validation Accuracy: {mean_acc:.4f} ± {std_acc:.4f}")

################################################################################
# 5) Train on all Train+Val, then evaluate on Test Set
################################################################################

# Build final model
final_model = create_transfer_model(input_shape=(128,128,3), fine_tune_at=100)

# Create dataset from entire trainval_df
ds_trainval = create_dataset(trainval_df, batch_size=BATCH_SIZE, shuffle=True)

# Compute class weights for entire trainval
trainval_counts = trainval_df["label"].value_counts()
benign_count = trainval_counts.get(0, 0)
malignant_count = trainval_counts.get(1, 0)
total = benign_count + malignant_count
weight_for_0 = (1 / benign_count) * (total / 2.0)
weight_for_1 = (1 / malignant_count) * (total / 2.0)
class_weights_final = {0: weight_for_0, 1: weight_for_1}

# Retrain on full trainval
final_history = final_model.fit(
    ds_trainval,
    epochs=EPOCHS,
    callbacks=callbacks,
    class_weight=class_weights_final,
    verbose=1
)

# Create dataset for test_df
ds_test = create_dataset(test_df, batch_size=BATCH_SIZE, shuffle=False)

# Evaluate final model on test set
test_loss, test_acc = final_model.evaluate(ds_test, verbose=0)
print(f"\n** Final Test Loss: {test_loss:.4f}, Final Test Accuracy: {test_acc:.4f}")

# Predictions & confusion matrix on test set
test_preds = final_model.predict(ds_test).ravel()
test_preds = (test_preds > 0.5).astype(int)

test_true = []
for _, lbl_batch in ds_test:
    test_true.extend(lbl_batch.numpy())

cm_test = confusion_matrix(test_true, test_preds)
print("\nConfusion Matrix (Test):\n", cm_test)
print("Classification Report (Test):\n",
      classification_report(test_true, test_preds,
                            target_names=["Benign (0)", "Malignant (1)"]))

# Optional: Plot the test confusion matrix
plt.figure(figsize=(5,4))
sns.heatmap(cm_test,
            annot=True, cmap='Blues', fmt='d',
            xticklabels=["Benign (0)", "Malignant (1)"],
            yticklabels=["Benign (0)", "Malignant (1)"])
plt.title("Confusion Matrix - Test Set")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()
```

```python
!pip install pynrrd

import os
import random
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras import models, layers
import nrrd
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns

################################################################################
# 0) Seed initialization for reproducibility
################################################################################
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)

################################################################################
# 1) Load the Excel file and create the DataFrame
################################################################################

excel_path = "/kaggle/input/my-lung-dataset/my_lung_dataset/dataset_lung.xlsx"
data_dir = "/kaggle/input/my-lung-dataset/my_lung_dataset/Train"

df = pd.read_excel(excel_path)

nodule_paths = []
labels_binary = []

for idx, row in df.iterrows():
    nodule_filename = row["Nodule"]
    tumor_class = row["TumorClass"]       
    full_nodule_path = os.path.join(data_dir, nodule_filename)
    
    # [1,2,3] => 0 (benign), [4,5] => 1 (malignant)
    label_bin = 0 if tumor_class in [1,2,3] else 1
    
    nodule_paths.append(full_nodule_path)
    labels_binary.append(label_bin)

data_nodules = pd.DataFrame({
    "nodule_path": nodule_paths,
    "label": labels_binary
})

print("Total nodule images:", len(data_nodules))
display(data_nodules.head())

################################################################################
# 1.5) First split: trainval (80%) + test (20%)
################################################################################

trainval_df, test_df = train_test_split(
    data_nodules, 
    test_size=0.20,
    stratify=data_nodules["label"],
    random_state=SEED
)

print(f"Train+Val samples: {len(trainval_df)}")
print(f"Test samples: {len(test_df)}")

################################################################################
# 2) Define image loading & dataset creation
################################################################################

IMG_SIZE = (128, 128)

def load_nrrd_image(nrrd_path, target_size=IMG_SIZE):
    volume_data, header = nrrd.read(nrrd_path)
    
    # If 3D, reduce to 2D if needed
    if volume_data.ndim == 3:
        if volume_data.shape[-1] == 1:
            volume_data = volume_data[..., 0]
        elif volume_data.shape[0] == 1:
            volume_data = volume_data[0, ...]
    
    volume_data = volume_data.astype(np.float32)
    volume_data = tf.image.resize(volume_data[..., tf.newaxis], target_size)
    volume_data = tf.image.grayscale_to_rgb(volume_data)  # 3 channels
    return volume_data.numpy()

def create_dataset(df, batch_size=8, shuffle=False):
    paths = df["nodule_path"].tolist()
    labels = df["label"].tolist()
    
    def gen():
        for p, l in zip(paths, labels):
            img = load_nrrd_image(p)
            yield img, l

    dataset = tf.data.Dataset.from_generator(
        gen,
        output_types=(tf.float32, tf.int32),
        output_shapes=((IMG_SIZE[0], IMG_SIZE[1], 3), ())
    )
    if shuffle:
        dataset = dataset.shuffle(buffer_size=len(df), seed=SEED)
    dataset = dataset.batch(batch_size).prefetch(tf.data.AUTOTUNE)
    return dataset

################################################################################
# 3) Create a Transfer Learning Model (DenseNet + Fine Tuning)
################################################################################

def create_transfer_model(input_shape=(128,128,3), fine_tune_at=None):
    base_model = tf.keras.applications.DenseNet121(
        input_shape=input_shape,
        include_top=False,
        weights='imagenet'
    )

    if fine_tune_at is None:
        base_model.trainable = False
    else:
        for layer in base_model.layers[:fine_tune_at]:
            layer.trainable = False
        for layer in base_model.layers[fine_tune_at:]:
            layer.trainable = True

    model = tf.keras.Sequential([
        base_model,
        layers.GlobalAveragePooling2D(),
        layers.Dense(128, activation='relu'),
        layers.Dropout(0.3),
        layers.Dense(1, activation='sigmoid')
    ])

    model.compile(
        optimizer='adam',
        loss='binary_crossentropy',
        metrics=['accuracy']
    )
    return model

################################################################################
# 4) Second split: train (e.g. 75% of trainval) + val (25% of trainval)
################################################################################
# Let's do a 75-25 split within trainval. Adjust as you like.

train_df, val_df = train_test_split(
    trainval_df,
    test_size=0.25,        # 25% val => effectively 60% train + 20% val, 20% test total
    stratify=trainval_df["label"],
    random_state=SEED
)

print(f"Train samples: {len(train_df)}")
print(f"Validation samples: {len(val_df)}")
print(f"Test samples: {len(test_df)}")

################################################################################
# 5) Train on train_df, validate on val_df
################################################################################

final_model = create_transfer_model(input_shape=(128,128,3), fine_tune_at=100)

ds_train = create_dataset(train_df, batch_size=8, shuffle=True)
ds_val   = create_dataset(val_df,   batch_size=8, shuffle=False)

# Compute class weights for train_df if needed
train_counts = train_df["label"].value_counts()
benign_count = train_counts.get(0, 0)
malignant_count = train_counts.get(1, 0)
total = benign_count + malignant_count
weight_for_0 = (1 / benign_count) * (total / 2.0)
weight_for_1 = (1 / malignant_count) * (total / 2.0)
class_weights_final = {0: weight_for_0, 1: weight_for_1}

# Callbacks
patience_es = 5
patience_lr = 3
callbacks = [
    tf.keras.callbacks.EarlyStopping(
        patience=patience_es, 
        restore_best_weights=True,
        monitor='val_loss',  # We'll monitor validation loss
        mode='min'
    ),
    tf.keras.callbacks.ReduceLROnPlateau(
        factor=0.1,
        patience=patience_lr,
        min_lr=1e-5,
        monitor='val_loss',
        mode='min'
    )
]

EPOCHS = 10
final_history = final_model.fit(
    ds_train,
    validation_data=ds_val,
    epochs=EPOCHS,
    callbacks=callbacks,
    class_weight=class_weights_final,
    verbose=1
)

################################################################################
# 6) Evaluate on the final test set
################################################################################

ds_test = create_dataset(test_df, batch_size=8, shuffle=False)

test_loss, test_acc = final_model.evaluate(ds_test, verbose=0)
print(f"\n** Final Test Loss: {test_loss:.4f}, Final Test Accuracy: {test_acc:.4f}")

# Predictions & confusion matrix on test
test_preds = final_model.predict(ds_test).ravel()
test_preds = (test_preds > 0.5).astype(int)

test_true = []
for _, lbl_batch in ds_test:
    test_true.extend(lbl_batch.numpy())

cm_test = confusion_matrix(test_true, test_preds)
print("\nConfusion Matrix (Test):\n", cm_test)
print("Classification Report (Test):\n",
      classification_report(test_true, test_preds,
                            target_names=["Benign (0)", "Malignant (1)"]))

plt.figure(figsize=(5,4))
sns.heatmap(cm_test,
            annot=True, cmap='Blues', fmt='d',
            xticklabels=["Benign (0)", "Malignant (1)"],
            yticklabels=["Benign (0)", "Malignant (1)"])
plt.title("Confusion Matrix - Test Set")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()
```
